# Types of Data

Everything carries data. Think about any object (physical, abstract, anything..). There can be attributes to this object.

Its relevant in the world of computer science, if the data can be collected(something that you can see), and brought to electronic format.

## First Classification of data

* Primary Data: Data that you(**researcher**) personally collected or made. The original form of data. e.g. Observations, Questionnaires, Group discussions, Interviews

* Secondary Data: Data that is not Primary. e.g. Census Data, Diaries, Web data

## Second Classification of data

### Quantitative

#### Discrete
e.g. Inventory. you could create bar chart with this data. This type of data looks like key-value pairs.
    
|Item|Quantity|      
|---|---|        
|Socks|10|        
|Shoes|5|        
|Footballs|15|           

#### Continuous

|Height
|---|
|176.5cm|
|177.4cm|
|183.0cm|
    
you could create a line chart with this data. This type of data will look like continuous change of a particular attribute with respect to something else.

### Qualitative
Such data is inherently discrete, as there can be only a finite number of categories

#### Nominal
e.g. Categories, pass/fail, red/blue/green, male/female
    
|Item|Category|
|---|---|
|Lipstick|Female|
|Hair Dryer|Female|
|Sanitary|Female|
|Jewellery|Female|
|Shaving Cream|Male|
|Hair Gel|Male|
|Wallet|Male|
    
you could create colored bubbles with red for male and yellow for female, pie chart
    
#### Ordinal
e.g. Ordered data, low/medium/high
    
|Item|Sale chances|
|---|---|
|Scarfs|High|
|Bags|High|
|Bracelets|Medium|
|Gowns|Low|
    
ordinal data have a relation in between unlike nominal data. e.g. red & blue are independent, they have no relationship. whereas, crimson/scarlet/ruby have an order in terms of amount of red in them.
    you could create stacked column chart

## Third Classification of data

### Interval
Data where difference matters

|Time|
|---|
|6:00  am|
|7:00  am|
|8:00  am|
|9:00  am|
|10:00 am|
|11:00 am|
|12:00 pm|
|13:00 pm|
|14:00 pm|
    
measure of scale with no absolute zero, you can operate on their difference, you can add the differences. but you can not get their ratios. you can never say that 12:00pm is twice of 6:00 am. These can be displayed in a line chart, bar chart

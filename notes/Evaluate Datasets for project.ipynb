```python
import pandas as pd
```

```python
data_url_df = pd.DataFrame(data={'Name':[
    '1000 movies data',
    'AB_test',
    'Breast_cancer_survival',
    'DecisionTree',
    'FastFoodRestaurants',
    'boxing_matches',
    'car_sales',
    'deliveries',
    'matches',
    'olympics dictionary',
    'summer olympics'
], 'URL': [
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/1000%20movies%20data.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/AB_test.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/Breast_cancer_survival.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/DecisionTree.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/FastFoodRestaurants.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/boxing_matches.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/car_sales.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/deliveries.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/matches.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/olympics%20dictionary.csv',
    'https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/summer%20olympics.csv'
], 'Encoding': [
    'utf-8',
    'utf-8',
    'utf-8',
    'utf-8',
    'utf-8',
    'utf-8',
    'iso-8859-1',
    'utf-8',
    'utf-8',
    'utf-8',
    'utf-8'
]})
```

```python
data_url_df['Rows'] = [0 for i in range(data_url_df.shape[0])]
data_url_df['Columns'] = [0 for i in range(data_url_df.shape[0])]
```

```python
data_url_df['DF'] = [pd.DataFrame() for i in range(data_url_df.shape[0])]
```

```python
data_url_df.set_index('Name', inplace=True)
```

```python
for i in range(data_url_df.shape[0]):
    dataSet = data_url_df.iloc[i]
    df = pd.read_csv(dataSet.URL, encoding=dataSet.Encoding)
    data_url_df.at[dataSet.name, 'Rows'] = df.shape[0]
    data_url_df.at[dataSet.name, 'Columns'] = df.shape[1]
    data_url_df.at[dataSet.name, 'DF'] = df
```
